```python
total_score = 0                                             # Part 2
options = ['A', 'B', 'C']


for item in input_list:

    if item[2] == "X":

        move = options[options.index(item[0])-1]            # Setzt 'move' auf Verlieremove -> ein Index nach links in 'options'
    
    elif item[2] == "Y":

        move = item[0]                                      # Setzt 'move' auf gegnerischen Move
    
    elif item[2] == "Z":

        if options.index(item[0]) < 2:

            move = options[options.index(item[0])+1]            # Setzt 'move' auf Gewinnermove -> ein Index nach rechts in 'options'. Edge-Case beachten zur Vermeidung von 'out of range' error.

        else:

            move = "A"

    score = 0

    if item[0] == 'A':

        if move == 'A':

            score += 4
        
        elif move == 'B':

            score += 8
        
        elif move == 'C':

            score += 3
    
    elif item[0] == 'B':

        if move == 'A':

            score += 1
        
        elif move == 'B':

            score += 5
        
        elif move == 'C':

            score += 9

    elif item[0] == 'C':

        if move == 'A':

            score += 7
        
        elif move == 'B':

            score += 2
        
        elif move == 'C':

            score += 6
    
    else:

        print ("Error: Wrong input")
        break
    
    total_score += score
```
